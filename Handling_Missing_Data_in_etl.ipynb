Q1. What are the most common reasons for missing data in ETL pipelines?
-- Missing data in ETL (Extract, Transform, Load) pipelines is very common, and it usually happens due to issues at different stages of data collection and processing. Here are the most common reasons:

1. Data Entry Errors
    .Users may skip fields or enter incomplete information.
Example: Optional form fields left blank.
2. System or Sensor Failures
    .Devices or systems may fail to capture data.
Example: IoT sensors going offline or malfunctioning.
3. Data Integration Issues
    .When combining multiple data sources, some fields may not match or exist in all sources.
Example: One database has “Phone Number” while another does not.
4. Schema Changes
    .Changes in database structure (adding/removing columns) can lead to missing values in older or newer data.
Example: A new column added but not populated for historical data.
5. Extraction Errors
    .Data may not be properly extracted due to API failures, connection issues, or query errors.
Example: API rate limits or timeout issues.
6. Transformation Errors
    .Incorrect transformation logic can accidentally remove or nullify values.
Example: Wrong join condition causing unmatched rows (NULL values).
7. Data Corruption
    .Files or records may get corrupted during transfer or storage.
Example: Broken CSV file leading to missing entries.
8. Intentional Missing Data
    .Some data is intentionally left blank.
Example: Users choosing not to disclose personal details (age, income).
9. Duplicate Removal or Filtering
    .While cleaning data, some values may be removed, creating gaps.
Example: Dropping duplicate rows without proper handling.
10. Timing Issues (Latency)
    .Data arrives late or is not available at the time of processing.
Example: Real-time pipeline missing delayed records.

Q3. Explain the difference between:
.Listwise deletion
.Column deletion
--
1. Listwise Deletion (Row Deletion)
Meaning: Remove entire rows (records) that contain any missing values.
How it works: If even one column in a row is NULL → the whole row is deleted.
Pros:

    1.Simple and easy to apply
    2.Keeps dataset clean (no missing values)
  Scenario:
In a dataset of 10,000 customers, only 50 rows have missing values → safe to delete those rows without affecting analysis.
2. Column Deletion
Meaning: Remove entire columns (features) that have too many missing values.
How it works: If a column has a high percentage of NULL values → drop that column.
Pros:

    1.Retains all rows (no loss of records)
    2.Useful when a column is mostly empty
  Scenario:
A “Middle Name” column is 80% empty → removing it won’t impact business insights.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?
-- Median imputation is preferred over mean imputation for skewed data (like income) because it is robust to extreme values (outliers) and better represents the typical value.

🔹 1. Effect of Skewness and Outliers
In skewed data (e.g., income), a few very high values can pull the mean upward.
The median is not affected by extreme values.

 Example:
Income = 10k, 12k, 15k, 18k, 5 lakh

Mean ≈ very high (distorted)
Median = 15k (more realistic)
🔹 2. More Representative of Typical Data
Median reflects the middle value, which is closer to what most people earn.
Mean may give an unrealistic value due to high-income outliers.
🔹 3. Preserves Data Distribution
Mean imputation can shift the distribution and reduce variability.
Median imputation keeps the distribution more natural and stable.
🔹 4. Reduces Bias
Using mean in skewed data introduces bias toward extreme values.
Median provides a balanced estimate.

Q5. What is forward fill and in what type of dataset is it most useful?
-- Forward Fill (FFill)

Forward fill is a method of handling missing data where a missing value is replaced with the last known (previous) non-missing value.
Forward fill is most useful in time-series datasets, where data is ordered over time.

2. Stock prices
3. Sensor readings (IoT data)
4. Weather data
5. Sales over time

Q6. Why should flagging missing values be done before imputation in an ETL workflow?
-- Flagging missing values before imputation is important because it preserves information about what was originally missing, which would otherwise be lost after filling the data.

🔹 1. Retains Original Information
Once you impute (fill) missing values, you can’t tell which values were originally NULL.
🔹 2. Improves Analysis & Modeling
    .Missingness itself can carry meaning.
    .Example: Customers who don’t provide income might behave differently.
🔹 3. Prevents Hidden Bias
    .Imputation may introduce artificial values.
    .Without a flag, these values look real and may mislead analysis.
🔹 4. Enables Better Decision-Making
    Analysts can choose to:
    .Use imputed values
    .Ignore flagged records
    .Treat them differently
🔹 5. Supports Data Auditing
    .Helps track data quality issues in ETL pipelines.
    .Useful for debugging and validation.
